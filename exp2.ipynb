```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image in grayscale
img = cv2.imread("/content/cat.jpg", 0)

# Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(img)

# Stack the original and CLAHE images side by side
res = np.hstack((img, clahe_img))

# Display the result
cv2_imshow(res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image in grayscale
img = cv2.imread("/content/cat.jpg", 0)

# Apply logarithmic transformation
c = 255 / np.log(1 + np.max(img))
log_img = c * (np.log(img + 1))

# Convert the result to 8-bit image
log_img = np.array(log_img, dtype=np.uint8)

# Stack the original and logarithmic transformed images side by side
res = np.hstack((img, log_img))

# Display the result
cv2_imshow(res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Read the image in grayscale
img = cv2.imread("/content/cat.jpg", 0)

# Apply gamma correction
gamma = 1.5
gamma_corrected = np.power(img / 255.0, gamma) * 255.0
gamma_corrected = np.array(gamma_corrected, dtype=np.uint8)

# Stack the original and gamma corrected images side by side
res = np.hstack((img, gamma_corrected))

# Display the result
cv2_imshow(res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
