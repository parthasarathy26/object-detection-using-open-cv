```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread("/content/6ab37dcdfd8150b6dfbb512b3b40e64b.jpg",cv2.IMREAD_GRAYSCALE)
plt.figure(1)
plt.imshow(image,cmap='gray')
plt.title("Orginal Image")
plt.show()
hist,bins=np.histogram(image.flatten(),256,[0,256])
thresh,_=cv2.threshold(image,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
otsu=cv2.threshold(image,thresh,255,cv2.THRESH_BINARY)[1]
plt.figure(2)
plt.imshow(otsu,cmap='gray')
plt.title("image segmentation with otsu thresholding")
plt.show()
```
